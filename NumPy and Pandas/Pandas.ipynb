## Pandas 
* Pandas is one of the most frequently using module in Python which is used for data analysis Purpose.
* Pandas deals with the tabular data by extrating its information, changing, merging, analysing it and etc.
* This tabular data may be in the form of .csv, .JSON, .excel and may be many more....

```python
#importing the pandas module and reading owi_
import numpy as np
import pandas as pd
covid_data = pd.read_csv("owid-covid-data.csv")
covid_data
```

```python
#Checking the columns in the present dataframe
covid_data.columns
```

```python
#Removing irrelevant fields from the data-frame
covid_df = covid_data[['iso_code', 'location', 'date', 'new_cases', 'new_deaths', 
                        'total_cases', 'total_deaths', 'hospital_beds_per_thousand',
                       'life_expectancy', 'human_development_index', 'population' ]].copy()
covid_df
```

```python
#Creating a new dataset which containing covid related information for the location "India" 
covid_india_df = covid_df[covid_df.location == "India"]
covid_india_df
covid_india_df = covid_india_df.fillna(covid_india_df.mean()) 
covid_india_df
```

```python
#Converting "date" column from objects to actual date using DateTime module
covid_india_df.date
covid_india_df['date'] = pd.to_datetime(covid_india_df.date)
covid_india_df.date
```

```python
#Converting date on the basis of year, month, day, weekday
covid_india_df["year"] = pd.DatetimeIndex(covid_india_df.date).year
covid_india_df["month"] = pd.DatetimeIndex(covid_india_df.date).month
covid_india_df["day"] = pd.DatetimeIndex(covid_india_df.date).day
covid_india_df["weekday"] = pd.DatetimeIndex(covid_india_df.date).weekday
covid_india_df
```

```python
#We can extract covid related information for any particular month, year by using the above dataframe
covid_india_df_may = covid_india_df[covid_india_df.month == 5]
covid_india_df_may

#Stats for any particular year for may month
covid_india_df_year_2020 = covid_india_df_may[covid_india_df_may.year == 2020]
covid_india_df_year_2020 #Further we cna check for any particular date

# Overall status of covid for an entire year.
covid_india_df_year = covid_india_df.groupby('year')[["new_cases", "new_deaths", 
                                                    "total_cases", "total_deaths"]].sum()
covid_india_df_year
```

```python
# Dropping the columns of year, day, weekday, month from covid_india_df
covid_india_df = covid_india_df.drop(["year", "month", "day", "weekday"], axis = 1)
covid_india_df
```

```python
#Calculating death rate on the basis of new cases and deaths 
# (neglecting life expextancy and population)
death_rate_covid = covid_india_df.new_deaths.sum() / covid_india_df.new_cases.sum()
print("The overall Death rate reported in India only due to Covid-19 is : {:.2f} %".format(death_rate_covid*100))
```

```python
#Calculating the cases and deaths per million
covid_india_df["cases_per_million"] = covid_india_df.total_cases*1e6 / covid_india_df.population
covid_india_df["deaths_per_million"] = covid_india_df.total_deaths*1e6 / covid_india_df.population
covid_india_df
```

```python
#Sorted the given databases on the basis of columns and saving it in the form of csv
covid_india_df
covid_india_df.to_csv("covid_india_df.csv", index=None)
```
